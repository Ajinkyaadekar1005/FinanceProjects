```python

import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt

tickers=['MIRZAINT.BO','HFCL.BO', 'MSTCLTD.NS', 'SJVN.BO', 'SUBEX.BO' ]
mydata=pd.DataFrame()
for t in tickers:
    mydata[t]=wb.DataReader(t,data_source='yahoo',start='2021-1-1')['Adj Close']
mydata.info()
```

```python
mydata.plot(figsize=(15,6));
```

Calculating simple returns of a portfolio with multiple securitites

```python
returns = (mydata/mydata.shift(1))-1
returns
weights = np.array([0.20,0.20,0.20,0.20,0.20])
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean()*250
annual_returns
```

```python
np.dot(annual_returns,weights)
portfolio_1 = str(round(np.dot(annual_returns,weights),5)*100)+"%"
```

FOR WEIGHT 2

```python
weights_2 = np.array([0.30,0.30,0.10,0.10,0.20])
portfolio_2 = str(round(np.dot(annual_returns, weights_2),5)*100) +'%'
```

```python
print(" return of portfolio 1 =" + portfolio_1)
print(" return of portfolio 2 =" + portfolio_2)
```
