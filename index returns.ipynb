```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt

tickers = ['^CNXIT', '^IXIC' ,'^BSESN']
indata=pd.DataFrame()
for t in tickers:
    indata[t]=wb.DataReader(t,data_source='yahoo',start='2010-1-1')['Adj Close']
indata.head()

(indata/indata.iloc[0]*100).plot(figsize=(15,6));
plt.show()
```

```python
ind_return=(indata/indata.shift(1))-1
print(ind_return)

annual_return=ind_return.mean()*250
print(annual_return)
```
